Opération de découpage: <br>
Découpez les deux premières lignes de données<br> Découpez les deux premières colonnes de données<br> Découpez les deux premières colonnes de données dans les deux premières lignes<br> Retournez les données du tableau<br> Exercice : retournez le haut, le bas, la gauche et la droite de une image<br> Exercice : couper la zone spécifiée de l'image

```python
#Découpez les deux premières lignes de données
arr
```

```python
arr[0:2]
```

```python
#Découpez les deux premières colonnes de données
arr[:,0:2] #arr[découpage des lignes, découpage des colonnes]
```

```python
#Découpez les deux premières colonnes de données dans les deux premières lignes
arr[0:2,0:2]
```

```python
#Retournez les données du tableau
arr
```

```python
#Inverser les lignes du tableau
arr[::-1]
```

```python
 #Inverser les colones du tableau
arr[:,::-1]
```

```python
#Tous les éléments inversés
arr[::-1,::-1]
```

Retourner une image à gauche et à droite

```python
import matplotlib.pyplot as plt
image = plt.imread('./1.jpeg')
plt.imshow(image)
```

```python
image.shape
```

image[ligne,colone,couleur]

```python
imageHorInv = image[:,::-1]
plt.imshow(imageHorInv)
```

Retourner une image en haut et en bas

```python
imageVerInv = image[::-1]
plt.imshow(imageVerInv)
```

Retourner tout le centre de l'image symétriquement

```python
imageInv = image[::-1,::-1]
plt.imshow(imageInv)
```

 Couper la zone spécifiée de l'image

```python
imageCoupé = image[0:100,0:100]
plt.imshow(imageCoupé)
```
