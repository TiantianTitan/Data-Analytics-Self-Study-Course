### 2.Numpy model

NumPy est la bibliothèque de base pour le calcul scientifique en langage Python. Il se concentre sur les calculs numériques et constitue également la base de la plupart des bibliothèques de calcul scientifique Python, principalement utilisées pour les opérations numériques effectuées sur des tableaux volumineux ou multidimensionnels.

#### La création de numpy

```python
import numpy as np
```

#### La création de la liste et le tableau

dimension 1 tableau

```python
arr = np.array([1,2,3])
arr
```

multi-dimensions tableau

```python
arr = np.array([[1,2,3],[4,5,6]])
arr
```

La différence entre le tableau et la liste:
Le type d'élément stocké dans le tableau doit être un type uniforme et la liste peut ne pas être uniforme.

```python
arr = np.array([1,2.2,"three"])
arr
```

Chargez une image externe dans un tableau numpy, puis essayez de modifier la valeur des éléments du tableau pour voir l'impact sur l'image d'origine.

```python
import matplotlib.pyplot as plt
img_arr = plt.imread('./1.jpeg') ## return ce tableau, c'est le contenant d'image
img_arr
plt.imshow(img_arr) #Visualiser les tableaux numpy
```

```python
img_arr = img_arr - 100 # Tous les éléments dans ce tableau moins 100
plt.imshow(img_arr)
```
