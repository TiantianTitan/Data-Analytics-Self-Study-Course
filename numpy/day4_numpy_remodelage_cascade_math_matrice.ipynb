### Matrice

Numpy contient une bibliothèque de matrices numpy.matlib, et les fonctions de ce module renvoient une matrice au lieu d'un objet ndarray. Une matrice est un tableau rectangulaire de lignes et de colonnes.Numpy contient une bibliothèque de matrices numpy.matlib, et les fonctions de ce module renvoient une matrice au lieu d'un objet ndarray. Une matrice est un tableau rectangulaire de lignes et de colonnes.

La fonction numpy.eye(n) renvoie la matrice d'identité d'une taille n. La matrice d'identité est une matrice carrée, et les éléments sur la diagonale du coin supérieur gauche au coin inférieur droit (appelée la diagonale principale) sont tous 1 et les autres éléments sont tous 0.

```python
np.eye(6)
```
